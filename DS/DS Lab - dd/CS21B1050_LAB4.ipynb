```python
#Question 5
'''
a.Display the length of the string and ask the user to provide a new paragraph if the length is less than 50.
b. Display the total number of words and sentences in the paragraph
c. Make the first character of each sentence uppercase.
d. Replace vowels with their next character. a is replaced by b, e by f etc.
e. Replace all S or s with $
'''
import string

text = input("Enter the paragraph: \n")
#Question A - length is less than 50
flag = True
while(flag == True):
    textlen = len(text)
    if textlen < 50:
        print(f"Length of the string is {textlen}, which is less than 50. Try again.\n")
        text = input("Enter the paragraph: \n") #Minimum 3 sentences
        flag = True
    else:
        print(" -> Passed test A")
        flag = False
       
#Question B - Total number of words and sentences in the paragraph
words = len(text.split())
print(f"-> The total number of words in the paragraph are {words}")
sentences = text.count(".")
print(f"-> The total number of sentences are {sentences}")

#Question C - First character of each sentence uppercase
textcaps = '. '.join([sentence.strip().capitalize() for sentence in text.split('.')])
print(f"-> Output after first character of each sentence is uppercase:\n {textcaps}\n")

#Question D - Replace vowels with their next character
replacements = {'a': 'b', 'A': 'B','e': 'f', 'E': 'F','i': 'j', 'I': 'J','o': 'p', 'O': 'P','u': 'v', 'U': 'V'}
    
vowtext = ''.join([replacements[char] if char in replacements else char for char in text])
print(f"-> The Replaced paragraph is:\n {vowtext}\n")

#Question E - Replace all S or s with $
replace = {'s': '$', 'S': '$'}
stext = ''.join([replace[char] if char in replace else char for char in text])
print(f"-> The Replaced paragraph S,s->$ is:\n {stext}\n")
```

```python
'''	
Question 1
Write a python program using classes to implement a stack with the following menu options, 1.Push, 2.Pop, 3.Peek, 4.Print, 5.Exit
'''	


def create():
    stack = []
    return stack
    
def empty(stack):
    return len(stack) == 0
    
def push(stack, item):
    stack.append(item)
    print(f"pushed item: {item}")

def pop(stack):
    if(empty(stack)):
        return "Stack in empty"
    return stack.pop()
    
def peek(stack):
    return stack[-1]


flag = True
stack = create()
while(flag):
    option = int(input("Enter the options,1)Push, 2)Pop, 3)Peek, 4)Print, 5)Exit: "))
    
    if option == 1:
        print("You have chosen push")
        
        #Push function
        ele = int(input("Enter the element: "))
        push(stack,ele)
        
        #Continue?
        yn =input("Continue[y/n]?: ")
        if yn == 'n':
            flag = False
        else:
            flag == True
        
    elif option == 2:
        print("You have chosen Pop")
        
        #pop function
        pop(stack)
        
        yn =input("Continue[y/n]?: ")
        if yn == 'n':
            flag = False
        else:
            flag == True
            
    elif option == 3:
        print("You have chosen Peek")
        
        #Peek function
        print(peek(stack))
        
        yn =input("Continue[y/n]?: ")
        if yn == 'n':
            flag = False
        else:
            flag == True
            
    elif option == 4:
        print("You have chosen Print")
        
        #Print
        print(stack)
        
        
        yn =input("Continue[y/n]?: ")
        if yn == 'n':
            flag = False
        else:
            flag == True
            
            
    elif option == 5:
        print("You have chosen Exit")
        flag = False
            
    else:
        print("Invalid input")
        yn =input("Continue[y/n]?: ")
        if yn == 'n':
            flag = False
        else:
            flag == True
```
