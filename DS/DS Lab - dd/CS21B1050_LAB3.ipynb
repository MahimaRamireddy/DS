```python
'''
Question 3
Get the items and its prices as follows (list of tuples) from the user and sort them according to their prices.
[(item1,price1),(item2,price2).....(item N ,price N)]
'''

def insert(items):
    item = input("Enter the item name: ")
    price = float(input("Enter the price: Rs."))
    itemtuple = (item,price)
    items.append(itemtuple)

def sorting(items):
    items.sort(key=lambda price: price[1])
            
    
items = []

stay = True

while stay != False:
    insert(items)
    x = input("Add more? [y/n]: ")
    if x == 'n':
        stay = False

print(f"The entered items are:\n{items}")
sorting(items)
print(f"The sorted items are:\n{items}")
```

```python
'''
Question 2
Get the quantity of items purchased and their prices from 5 users list as follows, Quantity=[Q1,Q2....Q5] Price=[p1,p2...p5].
a)	Compute the prices of the items purchased and return the answer as a single tuple.
Total amt=(TP1,TP2,TP,3,TP4,TP5).Where TP=Total amount of each person(Quantity * Price)
b)	Compute the total amount purchased by all the five users. 
'''

def insertquantity():
    quantity= []
    print("Quantity Value:")
    for i in range(0,5):
        x = int(input(f"Enter the quantity of {i+1} item: "))
        quantity.append(x)
    print(f"The quantity are: {quantity}")   
    return quantity
    
def insertprice():
    price= []
    print("Price Value:")
    for i in range(0,5):
        x = int(input(f"Enter the price of {i+1} item: "))
        price.append(x)
     
    print(f"The prices are: {price}")     
    return price
    
def total_amount(quantity,price):
    print("Question A)\t")
    print("The total amount is:")
    totalamount = []
    for i in range(0,5):
        totalamount.append(quantity[i] * price[i])
    print(totalamount)
    return totalamount

def total_amount_purchased(totalamount):
    print("Question B)\t")
    total = 0
    for i in range(0,5):
        total = total + totalamount[i]
    print(f"The total amount purchased by all the five users is: {total}")
        
    

q = insertquantity()
p = insertprice()
t = total_amount(q,p)
total_amount_purchased(t)
```

```python
'''
1.	Write a python script to implement student details using a dictionary The roll number of the student will be the key, the value will be a list containing the name, CGPA, and mobile number of the respective student. The program should be implemented as a menu-driven program with the following menus,
a.	Insert
b.	Delete
c.	Search
d.	Exit

'''
def insert(details,rollno,name,cgpa,mobileno):
    details[rollno] = [name,cgpa,mobileno]

def delete(details,rollno):
    del details[rollno]

def search(details,rollno):
    x =  details[rollno]
    return x
    
    
details = {}

stay = True
while stay!=False:
    option = int(input("Choose an Option:\n 1)Insert 2)Delete 3)Search 4)Exit : "))
    if option == 1:
        rollno = input("Enter the Roll Number: ")
        name = input("Enter the name: ")
        cgpa = float(input("Enter the CGPA: "))
        mobileno = int(input("Enter the Mobile Number: "))
        if (len(str(mobileno))) < 10:
            print("Invalid number, try again")
            break;
        insert(details,rollno,name,cgpa,mobileno)
        print(f"The details are:\n {details}")
        stay = True
    
    if option == 2:
        rollno = input("Enter the Roll Number to delete: ")
        delete(details,rollno)
        print(f"The details are:\n {details}")
        stay = True
        
    if option == 3:
        rollno = input("Enter the Roll Number to search: ")
        info = search(details,rollno)
        if info == None:
            print("Value doesn't exist")
        else:
            print(info)
        print(f"The details are:\n {details}")
        stay = True
        
    if option == 4:
        print("Thanks, Have a great day!")
        stay = False
```
