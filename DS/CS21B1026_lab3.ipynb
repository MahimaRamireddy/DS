```python
# q1


#function for inserting the values in dictionary
def insertion(dic,roll_no,name,cgpa,mobile_no):
    if roll_no not in dic:
        dic[roll_no]=[name,cgpa,mobile_no]
        print("Student information is added successfully.")
        print(dic[roll_no])
    else:
        print("Roll no already exists.")

    
#function for deleting the values in dictionary 
def deletion(dic,key):
    if key in dic:
        del dic[key]
        print("Student Info is deleted Successfully")
    else:
        print("Roll number not found !!")
        
#function for searching the values in dictionary         
def search(dic,key):
    if key in dic:
        print(dic[key])
    else:
        print("Roll number not found !!")
        
#creation of a dictionary to store student details 
info={}

#menu driven program
print("1.Insert")
print("2.Delete")
print("3.Search")
print("4.Exit")
a=int(input("Hello ! Please choose from the below options to continue :"))
while a!=4:
    if a==1:
        roll_no=input("Please enter the roll no of the student: ")
        name=input("Please enter the name of the student : ")
        cgpa=float(input("Please enter the cgpa of the student : "))
        if cgpa>=0 and cgpa<=10 :
            mobile_no=int(input("Please enter the mobile number of the student : "))
            if mobile_no>=1000000000 and mobile_no<=9999999999:
                insertion(info,roll_no,name,cgpa,mobile_no)
            else:
                print("Invalid phone number.The length of the phone number should be 10.")
        else:
            print("Invalid CGPA .The range of cgpa is between 0 and 10.")
            
        

        
    elif a==2:
        roll_no=input("Please enter the roll no of the student to delete: ")
        if info:
            deletion(info,roll_no)
        else:
            print("Dictionary is empty.")
    
    elif a==3:
        roll_no=input("Please enter the roll no of the student to search: ")
        if info:
            search(info,roll_no)
        else:
            print("Dictionary is empty.")
       
    a=int(input("Hello ! Please choose from the below options to continue :"))   
```

```python
#q2
quan=[]
pri=[]
mul=[]
error=False
for i in range(0,5):
    print("Person",i+1)
    a=float(input("Enter the quantity of the item purchased."))
    if a>0:
        quan.append(a)
        b=float(input("Enter the price of the item purchased."))
        if b>0:
            pri.append(b)
        else:
            print("The values shouldnt be negative.")
            error=True
            break  
    else:
        print("The values shouldnt be negative.")
        error=True
        break    
if not error:    
    print("The Quantities are :")
    print(quan)
    print("The prices are :")
    print(pri)

    for i in range(5):
        c=quan[i]*pri[i]
        mul.append(c)

    print("The product list is :")
    tuples=tuple(mul)
    print(tuples)

    add=sum(tuples)
    print("The total amount purchased by the five users is :")
    print(add)
```

```python
#q3

items=[]
pri=[]
error=False
n=int(input("Enter the number of users : "))
if n>0:
    for i in range(0,n):
        print("Person",i+1)
        a=input("Enter name of the item purchased.")
        items.append(a)
        b=float(input("Enter the price of the item purchased."))
        if b>0:
            pri.append(b) 
        else:
            print("The values shouldnt be negative.")
            error=True
            break
else:
    error=True
    
if not error:     
    print("The items are :")
    print(items)
    print("The prices are :")
    print(pri)

    tuples =list(zip(items,pri))
    print("The list of tuples is : ")
    print(tuples)

    sorted_list = sorted(tuples, key=lambda x: x[1])
    print("The sorted list is : ")
    print(sorted_list)
```
